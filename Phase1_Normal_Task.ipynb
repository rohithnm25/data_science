```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = ("/content/Hackathon_Ideal_Data.csv")
df = pd.read_csv(data)
```

```python
heat_map = df[['VALUE', 'QTY']].corr()
f, ax = plt.subplots(figsize = (8,4))
sns.heatmap(heat_map, vmax=0.5, square = True, annot = True)
```

```python
Line_plot = df.groupby('QTY')['VALUE'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(Line_plot['QTY'], Line_plot['VALUE'], marker='o', linestyle='--', color='black')
plt.title('Line chart of quantity vs value')
plt.xlabel('QTY')
plt.ylabel('VALUE')
plt.grid(True)
plt.show()
```

```python
pd.crosstab(index = df['STORECODE'],
            columns = df['VALUE'],
            margins = True,
            normalize = 'columns')
```

```python
sns.set(style="darkgrid")
sns.regplot(x=df['QTY'], y=df['VALUE'], color="black")
```

```python
V_counts = df['STORECODE'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(V_counts.index, V_counts, color='black')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
sns.distplot(df['VALUE'], color="black")
```

```python
sns.boxplot(x='QTY',data=df,color ="black")
```

```python
plt.figure(figsize=(10, 8))
sns.stripplot(x="GRP", y="VALUE", data=df, color="black")
plt.title('Strip Plot for Qty and Value')
plt.xlabel('Qty')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.stripplot(x="CMP", y="MBRD", data=df, color="black")
plt.title('Strip Plot for CMP and MBRD')
plt.xlabel('CMP')
plt.ylabel('MBRD')
plt.show()
```
